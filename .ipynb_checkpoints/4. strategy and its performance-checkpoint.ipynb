# portfolio performance



```python
def port_alpha_carhart(data_return,carhart,risk_free):
    data_return=data_return.drop(["2017/04"])
    carhart=carhart.drop(["2017/04"])
    risk_free=risk_free.drop(["2017/04"])
    data_return=data_return.sub(risk_free['Rf'], axis=0)
    endog = data_return
    exog = sm.add_constant(carhart[["RiskPremium","SMB","HML","UMD"]])
    rols = sm.OLS(endog, exog)
    rres = rols.fit()
    alpha= rres.params["const"].copy()
    return alpha

def port_alpha_ff(data_return,ff,risk_free):
    data_return=data_return.drop(["2017/04"])
    risk_free=risk_free.drop(["2017/04"])
    ff=ff.drop(["2017/04"])
    data_return=data_return.sub(risk_free['Rf'], axis=0)
    endog = data_return
    exog = sm.add_constant(ff[["RF","MKT","SMB","HML"]])
    rols = sm.OLS(endog, exog)
    rres = rols.fit()
    alpha= rres.params["const"].copy()
    return alpha


    
```

```python
def information_ratio(returns, factor_returns):
    if len(returns) < 2:
        return np.nan

    active_return = returns - factor_returns
    tracking_error = np.std(active_return, ddof=1)
    if np.isnan(tracking_error):
        return 0.0
    if tracking_error == 0:
        return np.nan
    return np.mean(active_return) / tracking_error
```

```python
def Strategy_performance(return_df,carhart,risk_free): 

    ser: pd.DataFrame = pd.DataFrame()
    ser['annual return'] = return_df.apply(lambda x: 12*np.mean(x))
    ser['sigma'] = return_df.apply(lambda x: np.std(x))
    ser['sharp'] = return_df.apply(lambda x: cal_sharp(x))
#     ser['IR'] = return_df.apply(
#             lambda x: information_ratio(x, carhart["RiskPremium"]))
    ser['alpha_Carhart'] = return_df.apply(lambda x: port_alpha_carhart(x,carhart,risk_free))
    ser['alpha_FF'] = return_df.apply(lambda x: port_alpha_ff(x,ff,risk_free))
    return ser.T
```
