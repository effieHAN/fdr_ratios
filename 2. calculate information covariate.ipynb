```python
cov=[rsquare,fund_size,return_gap,fund_flow,sharp,treynor,beta,vol]
combo=[]
pos_alpha=alpha[alpha['2020/03'].values > 0]
for i in cov:
    i=i[i.index.isin(pos_alpha.index)]
    i['rank'] = i['2020/03'].rank(ascending=False)
    i=i[["rank"]]
    combo.append(i)    
```

```python
combo=[alpha]+[p]+combo
combo=pd.concat(combo, axis=1)
combo.columns=["alpha",'p','rsquare','fund_size','return_gap','fund_flow','sharp','treynor','beta','vol']
combo=combo.drop(combo[combo.index=="sum"].index)
combo
```

```python
combo=combo.dropna()
combo
```
